# Methods .groupby() and .agg()

```python
import pandas as pd
```

```python
df_cars = pd.read_csv('Car_sales.csv')
df_cars
```

```python
df_cars.value_counts('Vehicle_type')
```

```python
df_cars = df_cars[['Manufacturer', 'Sales_in_thousands',
                   'Vehicle_type','Price_in_thousands', 
                   'Engine_size', 'Horsepower', 'Fuel_capacity']]

df_cars
```

```python
# a continuación vamos a explorar cuántos elementos únicos hay en el Dataframe usando el método unique()
df_cars.nunique()
# esta información es importante en el momento en el que realicemos agrupaciones
# también nos da informacion previa sobre la cantidad de datos que tendremos en las agrupaciones
```

## Method .agg()

```python
# las funciones agregadas son aquellas con las que podemos hacer operaciones con los datos
# la suma, media, resta etc.
```

```python
# comenzamos realizando la suma de las columnas
df_cars.agg('sum')
```

```python
df_cars.agg('mean')
```

```python
# si queremos realizar varios calculos a la vez, utilizamos de nuevo .agg pero abrimos lista [] para indicar las funciones
df_cars.agg(['sum', 'mean'])
```

```python
# a continuación aplicaremos dos funciones agregadas distintas a una misma columna.
# ahora seguiremos trabajando con listas, pero añadiremos diccionarios.
# se introduce diccionario {_:_} para indicar las diferentes funciones de agregación sobre una columna
# dentro del diccionario, introducimos una lista [] con las funciones agregadas a aplicar
df_cars.agg({'Sales_in_thousands':['sum','mean'], 
             'Price_in_thousands':['sum', 'max']})
```

```python
# ahora realizaremos la agregación a partir de las columnas.
# este caso añadimos el argumento axis=1 dentro del metodo .agg, para indicar que la agregación tendrá lugar sobre las columnas
df_cars[['Sales_in_thousands', 'Price_in_thousands']].agg('sum', axis=1)

# esta función lo que nos devuelve es:
# suma del valor en columna Sales_in_thousands y del valor Price_in_thousands por cada fila
```

```python
# también podemos hacer este calculo seleccionando cada una de las columnas por separado
# pero aquí hay una pequeña diferencia, ya que en las filas donde hay datos nulos, considera todo como nulo
# aunque realmente no todos los valores son nulos, solo un valor es nulo.
df_cars['Sales_in_thousands'] + df_cars['Price_in_thousands']

# por tanto esta función considera toda la suma como nulo si uno de los sumandos es nulo
# mientras que la función anterior considera los nulos como 0 y por tanto realiza la suma con un sumando
```

```python
df_cars
```

```python
# cómo renombrar el nombre de los index al hacer estas funciones de agregacion
# aquí usaremos de nuevo el metodo .agg pero con el formato tupples ()
df_cars.agg(x=('Sales_in_thousands', 'sum'),
            y=('Price_in_thousands', 'sum'))

# como podemos ver está agrupando los valores como nulos.
# para la agrupación correcta sin NaN, utilizaremos groupby
```

# Strategy Split-Apply-Combine

```python
# en primer lugar vamos a realizar un split del tipo de vehiculo car y passenger
df_cars['Vehicle_type'].value_counts()
```

## SPLIT

```python
# SPLIT
# esto es el split
# a continuacón definimos dos condiciones y le asignamos un valor a cada uno.
car_filter = df_cars['Vehicle_type']=='Car'
passenger_filter = df_cars['Vehicle_type']=='Passenger'
```

## APPLY

```python
# APPLY
# el siguiente paso a realizar es apply
# vamos a utlizar el split anterior para la agregación
df_cars[car_filter]
# se muestra el dataframe basado en el el tipo de condición definido.Vehicle_type=Car
```

```python
# para simplifcar el dataframe vamos a seleccionar una columna y realizar una operación
car_average = df_cars[car_filter]['Sales_in_thousands'].mean()
passenger_average = df_cars[passenger_filter]['Sales_in_thousands'].mean()
```

## COMBINE

```python
pd.DataFrame({'Vehicle_type':['Car', 'Passenger'],
             'Sales_in_thousand':[car_average, passenger_average]}).set_index('Vehicle_type')
# obtenemos una combinación entre el tipo de coche Car y Passsenger gracias al método Split+apply+combine
# indicamos .set_index para mostrar vehicle_type como index en lugar de como columna (probar a correr sin .set_index)
```

```python
# existe un método mas sencillo para realizar agrupaciones en pandas
# mediante el método groupby
```

# Method groupby

```python
# con este método podremos realizar el metodo split+apply+combine en una sola linea de codigo
```

```python
# nos permite agrupar los valores en categorias y aplicarles un cálculo
df_cars.groupby('Vehicle_type').mean()
```

```python
# si quisieramos que el index Vehicle_type apareciera como columna:
df_cars.groupby('Vehicle_type', as_index=False).mean()
```

```python
# a continuacion agruparemos por 'Manufacturer'
objeto_groupby= df_cars.groupby('Manufacturer')
# lo que genera no son datos en si, si no,un objeto, el cual podemos utilizar para otras operaciones
objeto_groupby
```

```python
# utilizando el atributo 'groups' podemos ver los elemento del agrupamiento bajo groupby
objeto_groupby.groups
# el rsultado lo obtenemos en forma de diccionario, mostrando como el index (valor de'Manufacturer) y los valores del resto de columnas
```

```python
# para mostrar solo las llaves
# obtenemos exclusivamente el nombre de todos los valores de 'Manufacturer'
objeto_groupby.groups.keys()
```

```python
# para observar los datos correspondientes a la agrupación 'Ford'
objeto_groupby.get_group('Ford')
```

## groupby().mean()

```python
# a continuación realizamos agrupación por 'Manufacturer' de la media de los datos
# ordenamos los valores de mayor a menor en función de la columna 'Sales_in_thousands'
df_cars.groupby('Manufacturer').mean().sort_values(by=['Sales_in_thousands'], axis=0, ascending=False)
```

## groupby().sum()

```python
# a continuación realizamos agrupación por 'Manufacturer' de la suma de los datos
# ordenamos los valores de mayor a menor en función de la columna 'Sales_in_thousands'
df_cars.groupby('Manufacturer').sum().sort_values(by=['Sales_in_thousands'], axis=0, ascending=False)
```

## groupby().count()

```python
df_cars.groupby('Vehicle_type').count().sort_values(by=['Sales_in_thousands'], axis=0, ascending=False)
```

```python
# en el dataframe podemos observar diferencias entre columna 'Sales_in_thousand' y 'Price_in_thousand'
# esta diferencia muestra que en una columna ha contado 41 y en otra 40.
# Esto puede ser debido a que existan valores nulos en dichas columnas y no han sido consideradas por count().
# lo comprobamos de la siguiente manera
```

```python
# observamos que efectivamente existen valores nulos en diferentes columnas
df_cars.isnull().sum()
```

```python
# a continuación vamos a mostrar esos valores nulos de las columnas
df_cars.groupby('Engine_size', dropna=False).count()
```

# Method filter() + groupby

```python
# Para filtrar valores en la función agregada suma.
# nos ayuda a filtrar en función de objetos agregados
# para ello debemos crear una funcion previamente donde x hace referncia a un DF.
# en return, se indica x, que es el DF seguido de [] que selecciona la columna para la que queremos filtrar
# dentro del metodo filter() introducimos el nombre asignado a la funcion def que es filter_func
```

```python
# en esta caso el DF mostrará Manufacturer cuya suma total de 'Sales_in_thounsands sea > 50
def filter_func(x):
    return x['Sales_in_thousands'].sum() > 52
df_filter = df_cars.groupby(['Manufacturer']).filter(filter_func)
df_filter
```

```python
# verificamos que manufacturers tienen suma sales <52
df_cars.groupby(['Manufacturer']).sum()['Sales_in_thousands'].sort_values().head(7)
```

```python
# comprobamos que con nuestra funcion filter, no aparecen estos 6 Manufacturers con suma <52
df_cars[df_cars['Manufacturer'].isin(['Porsche','Jaguar','Saab','Infiniti','Audi','BMW'])].head()
# a continuación calculamos el shape del df
```

```python
# con la funcion shape, obtenemos el numero de filas y columnaas
# el 13, indica el numero de Manufacturer que están presentes en el df
df_cars[df_cars['Manufacturer'].isin(['Porsche','Jaguar','Saab','Infiniti','Audi','BMW'])].shape
```

```python
# para saber cuántos Manufacturers habia en el df original hacemos:
print(df_cars.shape)
print(df_filter.shape)
# observamos que en df original habia 157 filas de Manufacturer
# y en df filtrado 144 filas de Manufacturer
# por tanto las 13 filas restantes son las que ya no aparecen en el df despues de filtrar
# ya que estas 13 son aquellas cuyo valor de suma <52
```
